```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import plotly.graph_objects as go
import plotly.express as px
from math import sqrt
import array as arr
from math import pi
import math
%matplotlib inline
```

```python
#Initialize variables
#Pr,Pt in mW 
#c is speed of light (m/s)
#f in MHz
#R in Km

Pr = 2*10**-14
Pt = 25
c = 3*10**8
f = 868*10**6
z = 6
L = 50
N = 1000
C1 = 5470
C2 = 3125
C3 = 1760
C4 = 980
C5 = 440
C6 = 250
Cap_Tot = C1+C2+C3+C4+C5+C6
CAP = [C1,C2,C3,C4,C5,C6]
DC = 1/100
Radius = []
Radius2 = []
N_Devices = []
N_Devices1 = []
N_Devices2 = []
S_tot = []
S_tot1 = []
S_tot2 = []
split_load = []
split_load1 = []
split_load2 = []
Pc = [0.454,0.26,0.146,0.08,0.036,0.021]
#Using Friis free space equation
#Pr = Pt* (c/4*pi*f)**2 * (1/R)**alpha
#On rearranging the equation
#R = ((Pt/Pr * (c/(4*pi*f))**2) **(1/alpha))

loss_exponent = np.arange(3,5,0.25)
    
for alpha in loss_exponent:
        R = ((Pt/Pr * (c/(4*pi*f))**2) **(1/alpha))/1000  #To get the value in Km
        #print("Max.Radius for alpha",str(alpha),"is:",round(R,3))
        A = pi * (R)**2
        #print("Total area for alpha",str(alpha), "is :",round(A,3))

#R1 = R2 = R3 = R4 = R5 = R6 = R/z   #Scenario 1 (EQUAL RADIUS)

        #print("Radius for each ring for alpha",str(alpha),":",round(R/z,3))
        Radius.append(round(R,3))
        Area = []
        

        for k in range(1,7):
            
                Ak = pi * ((k*R/z)**2 - ((k-1) * R/z)**2)
                #print("Area", str(k),"is:",round(Ak,3))
                Area.append(round(Ak,3))
                Nk = (N*Ak)/A
                #print("Number of devices in the Area",str(k),"for alpha",str(alpha),":",round(Nk,3))
                N_Devices.append(round(Nk,3))
    
    


        #Calculating packets sent 
        #Packets = [float(b)/float (m) for b, m in zip(CAP,N_Devices)]
        #print(Packets)

        size = L*8
        Pck_k = Cap_Tot/(N*size)
        #print("Packets/sec:",Pk)

        #Pk = [x/size for x in Packets]   #For individual SFs
        #load = [(float(b) * float(m))/int(n) for b, m, n in zip(N_Devices,Pk,BR)]    #For individual SFs
        #print(load)
        #offered_load = [x * size * DC for x in load]    #For individual SFs

        G = (N * Pck_k * size * DC )/ Cap_Tot
        split_load = [b * G for b in (Area)]
        #print("G for each SF:",split_load)

        S = [i * math.exp(-2*i) for i in (split_load)]
        #print("Throughput:",S)

        throughput = sum(S) 
        S_tot.append(round(throughput,3))

'''        plt.figure(figsize=(8,4))
        plt.plot(split_load,S,'bo',linestyle='solid')
        plt.xlabel("G")
        plt.ylabel("S")
        plt.show()
'''

print(S_tot)

plt.figure(figsize=(8,4))
plt.plot(loss_exponent,S_tot,color='black',marker='o',linestyle='solid')
plt.xlabel("loss exponent")
plt.ylabel("Throughput(bps)")
plt.show()
```

```python
#A1 = A2 = A3 = A4 = A5 = A6 = A/z   #Scenario 2 (EQUAL Area)

for alpha in loss_exponent:
        R = ((Pt/Pr * (c/(4*pi*f))**2) **(1/alpha))/1000  #To get the value in Km
        #print("Max.Radius for alpha",str(alpha),"is:",round(R,3))
        A = pi * (R)**2
        #print("Total area for alpha",str(alpha), "is :",round(A,3))

        Area1 = []
        
        for m in range(1,7):
          Am = A/z
          #print("Area", str(m),"is:",round(Am,3))
          Area1.append(round(Am,3))
          Nm = (N*Am)/A
          #print("Number of devices in the Area",str(m),"for alpha",str(alpha),":",round(Nm,3))
          N_Devices1.append(round(Nm,3))
        
        size = L*8
        Pck_m = Cap_Tot/(N*size)
        #print("Packets/sec:",Pm)

        G1 = (N * Pck_m * size * DC )/ Cap_Tot
        split_load1 = [b * G1 for b in (Area1)]
        #print("G for each SF:",split_load1)

        S1 = [i * math.exp(-2*i) for i in (split_load1)]
        #print("Throughput:",S1)

        throughput1 = sum(S1) 
        S_tot1.append(round(throughput1,3))

print(S_tot1)

plt.figure(figsize=(8,4))
plt.plot(loss_exponent,S_tot1,color='black',marker='o',linestyle='solid')
plt.xlabel("loss exponent")
plt.ylabel("Throughput(bps)")
plt.show()
```

```python
# SCENARIO 3 (OPTIMAL CAPACITY)

for alpha in loss_exponent:
        R = ((Pt/Pr * (c/(4*math.pi*f))**2) **(1/alpha))/1000  #To get the value in Km
        #print("Max.Radius for alpha",str(alpha),"is:",round(R,3))
        A = (math.pi * (R)**2)
        #print("Total area for alpha",str(alpha), "is :",round(A,3))

        Radius2 = []
        Area2 = []
        N_Devices2 = []
        temp_k=0
        Pn = [x_a * (A/math.pi) for x_a in (Pc)] 
        Cn = np.sqrt(Pn)
        #print(Pn)
        #print(Cn)

        for x_b in Cn:
          #print(x_b)
          Rn=x_b-temp_k
          #print(Rn)
          Radius2.append(Rn)
          temp_k=temp_k+Rn
        #print(Radius2)
          

        for n in Radius2:
           An = math.pi * n * n
           Area2.append(An)
           Nn = (N * An)/A
           N_Devices2.append(Nn)
        #print(N_Devices2) 

        
        size = L*8
        Pck_n = Cap_Tot/(N*size)
        G2 = (N * Pck_n * size * DC )/ Cap_Tot
        split_load2 = [b * G2 for b in (Area2)]
        #print(split_load2)
        S2 = [i * math.exp(-2*i) for i in (split_load2)]
        #print("Throughput:",S2)
        throughput2 = sum(S2) 
        S_tot2.append(round(throughput2,3))

print(S_tot2)
plt.figure(figsize=(8,4))
plt.plot(loss_exponent,S_tot2,color = 'black',marker = 'o',linestyle = 'solid')
plt.xlabel("loss exponent")
plt.ylabel("Throughput(bps)")
plt.show()
```
